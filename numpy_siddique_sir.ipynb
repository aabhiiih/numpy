# **EYE METHOD**

the **eye tool returns a 2-D array with 1's as the diagonal and 0's elsewhere.** The diagonal can be main,upper,or lower depending on the optional parameter k. **A positive k is for the upper diagonal,a negative k is for the lower,and a 0k(defualt) is for the main diagonal.**

```python
# Python Programming illustratrating
# numpy.eye method

import numpy as np

# 2x2 matrix with 1's on main diagonal
b=np.eye(2)
print("Matrix b :\n",b)
print("--------------------")
b=np.eye(3)
print("Matrix b :\n",b)
print("-----------------")

#matrix with R=4 C=5 and 1 on diagonal
#below main diagonal
a=np.eye(4,5,k=-1)
print("\nMatrix a :\n",a)
```

```python
a=np.eye(4,5,k=1)
print("\nMatrix a :\n",a)
```

## ZIPF DISTRIBUTION

Zipf distribution are **used to sample data based on zipf's law.**


---


**Zipf's Law:**
 *  In a collection, the nth common term is 1/n times of the most common term. Eg. the 5th most common word in English occurs nearly 1/5 times as often as the most common word.
---
**It has two parameters:**
  *  **a** - distribution parameter.
  *  **size** - The shape of the returned array.
  
    The **Zipf distribution,** also known as Zipf's law, is a discrete probability distribution that follows a power-law pattern. In other words, the most frequent event occurs much more often than the second most frequent event which in turn occurs more often than the third, and so on.
    
    It is commonly observed in many natural, social, and linguistic phenomena, where a few occurrences are very common, and many occurrences are rare.

**Real-World Applications of the Zipf Distribution**

 **1. Word Frequencies in Natural Language**

 One of the classic examples of Zipf's law is the distribution of word frequencies in natural languages. A few words, such as "the", "of", and "and", are used very frequently, while the majority of words are used only rarely.

 **Example:** Word Frequency in English Texts

  *  In a typical English text, the most frequent word, like "the", occurs much more often than the second most frequent word, such as "be". The distribution of word frequencies follows a Zipf distribution, where a small number of common words dominate, and many words are used only a few times.

# **NUMPY UFUNCS**

**What are ufuncs?** •
- ufuncs stands for "Universal Functions" and they are NumPy functions that operate on the ndarray object.

**why use ufuncs?**
-  ufuncs are used to implement vectorization in NumPy which is way faster than iterating over elements.
- They also provide broadcasting and additional methods like reduce, accumulate etc. that are very helpful for computation.

**ufuncs also take additional arguments like:**
- where boolean array or condition defining where the operations should take place.
-  dtype defining the return type of elements.
-  out output array where the return value should be copied

```python
x=[1,2,3,4]
y=[4,5,6,7]
z=[]
for i,j in zip(x,y):
  z.append(i+j)
print(z)
```

```python
# with ufunc, we can use the add()

import numpy as np

a=[1,2,3,4]
b=[5,6,7,8]
c=np.add(a,b)
print(c)
print(type(np.add))
```

# **NUMPY GCD GREATEST COMMON MULTIPLE**

**Finding GCD(Greatest Common Denominator),also known as HCF (Highest Common Factor) is the biggest number that is a common factor of both of the numbers.**

```python
# Find the HCF of the foolowing two numbers:

import numpy as np

num1=6
num2=9

x=np.gcd(num1,num2)

print(x)
```

**Returns: 3 because that is the highest number both numbers can be divided by (6/3=2 and 9/3=3)**

---

 *  The np.gcd() function is used to compute the GCD of num1 and num2. The GCD is the largest integer that divides both numbers without leaving a remainder.
 *  In this case,the GCD of 6 and 9 is 3 is the largest number that divides both 6 and 9.

 ---

 *  **Explanation of GCD(Greatest Common Denominator):**

    *  For 6 and 9:
      *  The divisors of 6: 1, 2, 3, 6
      *  The divisors of 9: 1, 3, 9
      *  The common divisors: 1 and 3
      *  The largest common divisor is 3, which makes the GCD of 6 and 9 equal to 3.

## **Finding GCD in Arrays**

*  To find the Highest Common Factor of all values in an array, you can use the **reduce()** method.
*  **The reduce() method will use the ufunc, in this case the god() function, on each element, and reduce the array by one dimension.**


```python
# Find the GCD for all of the numbers in the following array

import numpy as np

arr=np.array([20,8,32,36,16])

x=np.gcd.reduce(arr)

print(x)
```

**Returns: 4 because that is the highest number all values can be divided by**

---
 *  The function **np.gcd.reduce(arr)** computes the GCD of the entire array by applying the GCD function cumulatively across all elements in the array.

 *  The reduce function applies the GCD operation iteratively:
     *  It first finds the GCD of the first two numbers (gcd(20, 8) = 4).
     
     *  Then, it takes this result and applies the GCD with the next number (gcd(4, 32) 4), and so on.
     
     *  This process continues **until all the numbers in the array are reduced to their greatest common divisor.**

---

*  Detailed GCD Calculation:
    *  Finding the GCD of the first two numbers:
    *  gcd(20, 8) = 4
       *  Finding the GCD of the result with the next number:
    *  gcd(4, 32) = 4
       *  Next:
    *  gcd(4, 36) = 4
        * Final Step:
    *  gcd(4, 16) = 4
  
*  The GCD of all the numbers in the array [20, 8, 32, 36, 16] is 4.

## What is a Set
---
*  A set in mathematics is a collection of unique elements
*  Sets are used for operations involving frequent **intersection, union and difference operations.**

## Create Sets in Numpy

---
*  We can use NumPy's **unique()** method to find unique elements from any array. Eg. create a set array, but remember that the set arrays should only be 1-D arrays.

```python
# Convert following array with repeated elements to a set:

import numpy as np

arr=np.array([1,1,1,2,3,4,5,5,6,7])

x=np.unique(arr)

print(x)
```

## Finding Union

---
*  To find the unique values of two arrays, use the **union1d() method.**

```python
# Find union of the following two set arrays:

import numpy as np

arr1=np.array([1,2,3,4])
arr2=np.array([3,4,5,6])

newarr=np.union1d(arr1,arr2)

print(newarr)
```

## Finding Intersection

*  To find only the values that are present in both arrays,use the **intersect1d() method.**

```python
# Find intersection of the following two set arrays:

import numpy as np

arr1=np.array([1,2,3,4])
arr2=np.array([3,4,5,6])

newarr=np.intersect1d(arr1,arr2,assume_unique=True)

print(newarr)
```

**Note: the intersect1d() method takes an optional argument assume_unique, which if set to True can speed up computation. It should always be set to True when dealing with sets.**

## Finding Difference

---
*  To find only the values in the first set that is NOT present in the seconds set, use the **setdiff1d() method.**

```python
# Find the difference of the set1 from set2:

import numpy as np

set1=np.array([1,2,3,4])
set2=np.array([3,4,5,6])

newarr=np.setdiff1d(set1,set2,assume_unique=True)

print(newarr)
```

Note: the **setdiff1d() method takes an optional argument assume_unique,** which if set to **True can speed up computation.** It should always be set to True when dealing with sets.

## Finding Symmetric Difference

*  To find only the values that are NOT present in BOTH sets, use the **setxor1d() method.**

```python
# Find the symmetric difference of the set1 and set2:

import numpy as np

set1=np.array([1,2,3,4])
set2=np.array([3,4,5,6])

newarr=np.setxor1d(set1,set2,assume_unique=True)

print(newarr)
```
